```python
query = '''

    SELECT
        ORIGIN
        ,COUNT(ORIGIN) AS CNT
    FROM
        covid_march_delays
    WHERE
        CANCELLED = 1
    GROUP BY
        ORIGIN
    ORDER BY 
        CNT DESC

'''

pd.read_sql(query, engine)
```

```python
query1 = '''

    SELECT
        OP_CARRIER
        ,COUNT(OP_CARRIER) AS CNT
    FROM
        march_predictions
    WHERE
        CANCELLED = 1
    GROUP BY
        OP_CARRIER
    ORDER BY 
        CNT DESC

'''

pd.read_sql(query1, engine)
```

```python
query2 = '''

    SELECT
        DAY_OF_MONTH
        ,COUNT(DAY_OF_MONTH) AS CNT
    FROM
        march_predictions
    WHERE
        CANCELLED = 1
    GROUP BY
        DAY_OF_MONTH
    ORDER BY 
        CNT DESC

'''

pd.read_sql(query2, engine)
```
